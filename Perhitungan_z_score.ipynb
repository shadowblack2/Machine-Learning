```python
#import library
import pandas as pd
from pygrowup import Calculator
from google.colab import files
import numpy as np
import io
```

```python
print("Silakan unggah file CSV 'Data Stunting dan Malnutrisi' Anda...")
uploaded = files.upload()
```

```python
# Dapatkan nama file yang diunggah dan baca ke dalam DataFrame
file_name = next(iter(uploaded))
print(f"\nFile '{file_name}' berhasil diunggah.")
df = pd.read_excel(io.BytesIO(uploaded[file_name]))
```

```python
print("\nMemulai perhitungan Z-score...")

# Inisialisasi kalkulator pygrowup
calculator = Calculator(
    adjust_height_data=False,
    include_cdc=True
)
```

```python
# List penampung hasil
haz_scores = []
waz_scores = []
whz_scores = []
stunting_list = []
wasting_list = []

print("Memulai perhitungan Z-score, status gizi...")

for index, row in df.iterrows():
    jenis_kelamin = row['Jenis Kelamin']
    age_in_months = row['Umur (bulan)']
    height = row['Tinggi Badan (cm)']
    weight = row['Berat Badan (kg)']

    # pengganti gender_mapped
    if jenis_kelamin == 'Laki-laki':
        gender = 'm'
    else:
        gender ='f'

    # Cek data kosong
    if pd.isna(height) or pd.isna(weight) or pd.isna(age_in_months) or pd.isna(jenis_kelamin):
        print(f"Baris {index} dilewati: data tidak lengkap")
        haz_scores.append(np.nan)
        waz_scores.append(np.nan)
        whz_scores.append(np.nan)
        stunting_list.append(np.nan)
        wasting_list.append(np.nan)
        continue

    try:
        # Pastikan tipe data numerik
        age_in_months = float(age_in_months)
        height = float(height)
        weight = float(weight)

        # Hitung HAZ & WAZ
        haz = calculator.lhfa(height, age_in_months, gender)
        waz = calculator.wfa(weight, age_in_months, gender)

        # Hitung WHZ (Weight-for-Length/Height)
        if age_in_months < 24:
            adjusted_height = height + 0.7  # konversi ke panjang tidur
            whz = calculator.wfl(weight, age_in_months, gender, adjusted_height)
        else:
            whz = calculator.wfh(weight, age_in_months, gender, height)

        # Simpan skor
        haz_scores.append(haz)
        waz_scores.append(waz)
        whz_scores.append(whz)

        # === Klasifikasi(HAZ) ===
        if haz < -3:
            stunting = "Severely Stunted"
        elif -3 <= haz < -2:
            stunting = "Stunted"
        else:
            stunting = "Normal"

        # === Klasifikasi (WHZ) ===
        if whz < -3:
            wasting = "Severely Wasted"
        elif -3 <= whz < -2:
            wasting = "Wasted"
        elif whz > 3:
            wasting = "Obese"
        elif whz > 2:
            wasting = "Overweight"
        else:
            wasting = "Normal"

        # === Klasifikasi (WAZ) ===
        if waz < -3:
            wasting= "Severely Underweight"
        elif -3 <= waz < -2:
            wasting = "Underweight"

        # Simpan hasil ke list
        stunting_list.append(stunting)
        wasting_list.append(wasting)

    except Exception as e:
        print(f"Baris {index} gagal dihitung: {e}")
        haz_scores.append(np.nan)
        waz_scores.append(np.nan)
        whz_scores.append(np.nan)
        stunting_list.append(np.nan)
        wasting_list.append(np.nan)

# Tambahkan hasil ke DataFrame
df['HAZ'] = haz_scores
df['WAZ'] = waz_scores
df['WHZ'] = whz_scores
df['Stunting'] = stunting_list
df['Wasting'] = wasting_list

# Hapus baris yang tidak punya hasil
df.dropna(subset=['HAZ', 'WAZ', 'WHZ', 'Stunting', 'Wasting'], inplace=True)

# Simpan ke file baru
output_filename = 'data_lengkap.csv'
df.to_csv(output_filename, index=False, sep=';')

print(f" Perhitungan selesai. Hasil disimpan sebagai '{output_filename}'.")
```
